## Objective
The main objective of this assignment is to simulate the motion of a damped driven pendulum numerically using the fourth-order Runge-Kutta (RK4) method. This project demonstrates how computational techniques solve nonlinear differential equations describing real-world physical systems.

## Theory and Background
A damped driven pendulum is governed by the differential equations:
dθ/dt = ω  
dω/dt = b*cos(ω₀*t) - q*ω - sin(θ)  

where θ is the angular displacement, ω is angular velocity, b*cos(ω₀*t) is the driving torque, and q*ω is the damping term. Analytical solutions are difficult, so numerical methods like RK4 are applied.

## Problem Statement
The task is to compute θ(t) and ω(t) for 50 seconds using RK4, given initial conditions θ=3 rad, ω=0 rad/s, and constants q=0.2, b=0.1, ω₀=1.

## Methodology
We use the fourth-order Runge-Kutta method for solving coupled differential equations. At each time step, intermediate slopes k1, k2, k3, k4 and c1, c2, c3, c4 are computed for θ and ω, then combined to update values.

## Calculations
The RK4 formulas are:  
θ_{i+1} = θ_i + (k1 + 2*k2 + 2*k3 + k4)/6  
ω_{i+1} = ω_i + (c1 + 2*c2 + 2*c3 + c4)/6


```python
import numpy as np
from math import cos, sin
import matplotlib.pyplot as plt

def f(t, theta, omega):
    return omega

def g(t, theta, omega):
    q = 0.2
    b = 0.1
    omega_0 = 1
    return b*cos(omega_0*t) - q*omega - sin(theta)

step_size = 0.001
t_initial = 0
t_final = 50

theta = [3]
omega = [0]

t = np.arange(t_initial, t_final + step_size, step_size)

for i in range(len(t) - 1):
    k1 = step_size*f(t[i], theta[i], omega[i])
    c1 = step_size*g(t[i], theta[i], omega[i])

    k2 = step_size*f(t[i] + step_size/2, theta[i] + k1/2, omega[i] + c1/2)
    c2 = step_size*g(t[i] + step_size/2, theta[i] + k1/2, omega[i] + c1/2)

    k3 = step_size*f(t[i] + step_size/2, theta[i] + k2/2, omega[i] + c2/2)
    c3 = step_size*g(t[i] + step_size/2, theta[i] + k2/2, omega[i] + c2/2)

    k4 = step_size*f(t[i] + step_size, theta[i] + k3, omega[i] + c3)
    c4 = step_size*g(t[i] + step_size, theta[i] + k3, omega[i] + c3)

    theta.append(theta[i] + (k1 + 2*k2 + 2*k3 + k4)/6)
    omega.append(omega[i] + (c1 + 2*c2 + 2*c3 + c4)/6)

plt.plot(t, theta, label='Position (angle) vs time')
plt.plot(t, omega, label='Angular velocity vs time')
plt.grid()
plt.legend()
plt.show()
```
