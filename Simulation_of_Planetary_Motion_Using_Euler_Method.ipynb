### Simulation of Planetary Motion Using Euler Method

## Objective
The primary objective of this assignment is to simulate the motion of a planet orbiting a star using the Euler method, a fundamental numerical approach for solving ordinary differential equations. This study aims to model the planet’s trajectory under the influence of Newtonian gravitational forces, visualize the orbit, and understand the limitations and accuracy of the Euler method in long-term simulations. By implementing this computational experiment, the assignment highlights the application of numerical techniques in solving real-world physics problems that are analytically intractable.

## Theory and Background
The motion of celestial bodies, such as planets around a star, is governed by Newton’s Law of Gravitation:  

\[
F = \frac{GMm}{r^2}
\]

where \(G\) is the universal gravitational constant, \(M\) is the mass of the star, \(m\) is the mass of the planet, and \(r\) is the distance between them.  

From Newton’s second law:

\[
a = \frac{F}{m} = - \frac{GM}{r^3} \vec{r}
\]

Analytical solutions for orbital motion are complex, particularly for non-circular or perturbed orbits. Numerical methods such as the Euler method provide a stepwise approximation.


## Problem Statement
We simulate a planet initially located \(1.496 \times 10^{11}\) m from the star, moving with an initial velocity of 29,780 m/s perpendicular to its position vector.  

- Simulation steps = 5000  
- Each step = 6 hours  
- Method = Euler integration  


## Methodology
1. Define constants: \(G, M\).  
2. Set initial conditions: position and velocity.  
3. At each step, calculate acceleration components:  

\[
a_x = -\frac{GMx}{(x^2+y^2)^{3/2}}, \quad
a_y = -\frac{GMy}{(x^2+y^2)^{3/2}}
\]

4. Update velocity:  

\[
v_x^{new} = v_x + a_x \Delta t, \quad v_y^{new} = v_y + a_y \Delta t
\]

5. Update position:  

\[
x^{new} = x + v_x \Delta t, \quad y^{new} = y + v_y \Delta t
\]

6. Store trajectory points and plot the orbit.  


```python
# Python Implementation

import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11   # Gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30      # Mass of the star (kg)
dt = 60 * 60 * 6  # Time step: 6 hours in seconds
steps = 5000      # Number of steps

# Initial conditions
x, y = 1.496e11, 0   # Initial position (m)
vx, vy = 0, 29780    # Initial velocity (m/s)

# Function to calculate gravitational acceleration
def acceleration(x, y):
    r = np.sqrt(x**2 + y**2)
    ax = -G * M * x / r**3
    ay = -G * M * y / r**3
    return ax, ay

# Lists to store positions
X, Y = [], []

# Euler integration loop
for _ in range(steps):
    ax, ay = acceleration(x, y)
    vx += ax * dt
    vy += ay * dt
    x += vx * dt
    y += vy * dt
    X.append(x)
    Y.append(y)

# Plot the orbit
plt.figure(figsize=(8,8))
plt.plot(X, Y, label="Planet Orbit (Euler Method)")
plt.plot(0, 0, 'yo', markersize=12, label="Star")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Planetary Motion Simulation using Euler Method")
plt.legend()
plt.axis("equal")
plt.grid()
plt.show()
```

## Discussion
- The Euler method is simple and effective for short-term approximations.  
- Errors accumulate over time, making the orbit unstable in long simulations.  
- For higher accuracy, methods such as **Runge-Kutta** should be used.

The experiment demonstrates the application of the Euler method to planetary motion.  
It shows that while qualitative orbital motion is captured, long-term stability is not guaranteed.
